```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline
```

```python
df= pd.read_csv("menu.csv")
```

```python
df
```

```python
df.columns
```

```python
stats_df= df[['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']]
```

```python
stats_df
```

```python
stats_df.head(10)
```

```python
stats_df.info()
```

```python
stats_df.describe()
```

```python
corr_with_calories= stats_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_with_calories, annot=True, fmt=".2f", )
plt.title("Correlation between Calories and other independent variables")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x="Calories", y= "Category", data=df, hue="Category")
plt.title("Boxplot for Calories vs Category")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.stripplot(x="Category", y="Calories", data=df,
              hue="Category", jitter=True)
plt.xticks(rotation=45)
plt.title("Calories Content")
plt.show()
```
